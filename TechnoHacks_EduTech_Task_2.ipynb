```python
# import all required libaries
import pandas as pd
import numpy as np

```

```python
#READ THE DATSET: Here I ve used Titanic dataset
data = pd.read_csv("train.csv")
```

```python
#First Srows of dataset
data.head()
```

```python
#Check for data type
data.dtypes
```

```python
# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns.tolist()

# Identify numerical columns
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
print ("Categorical columns:")
print (categorical_columns)
print ("Numerical columns:")
print (numerical_columns)
```

```python
# List of columns to be used
columns_used = ['PassengerId', 'Age', 'Survived', 'Pclass', 'SibSp', 'Parch', 'Fare']

# Select the specified columns from the DataFrame
selected_columns = data.loc[:, columns_used]
```

```python
selected_columns. head(5)
```

```python
column_means = selected_columns.mean()
print("Mean for each column:", column_means )
```

```python
# Calculate and print the mode for each column
column_modes = selected_columns.mode().iloc[0]
print("Modes for each column:", column_modes)
```

```python
# Calculate and print the median for each column
column_median = selected_columns.median()
print("Median for each column:", column_median)
```

```python
# Calculate and print the standard deviation for each column
column_std_deviation = selected_columns.std()
print("Standard Deviation for each column:", column_std_deviation)
```

```python
# Get summary statistics for each numeric column
summary_statistics = selected_columns.describe()
print("Summary statistics for each numeric column:\n", summary_statistics)
```
